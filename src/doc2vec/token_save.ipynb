# Tokenizer

```python
import swifter
import pandas as pd
from konlpy.tag import Okt
```

```python
jvm_path = "/Library/Java/JavaVirtualMachines/zulu-15.jdk/Contents/Home/bin/java"
okt = Okt(jvmpath=jvm_path)
```

```python
def tokenizer(content):
    result=okt.pos(str(content))
    return result
```

```python
df['token'] = df.content.swifter.apply(tokenizer)
```

```python
df = df[['content', 'token']]
```

```python
df.head()
```

```python
result = []
for content in zip(df.content, df.token):
    for i in content[1]:
        temp = []
        temp.append(content[0])
        temp.append(i)
        result.append(temp)
```

```python
result
```

```python
df_result = pd.DataFrame(result, columns=['origin', 'token'])
```

```python
df.shape
```

```python
df_result.head()
```

```python
df_result['token1'] = df_result.token.apply(lambda x: x[0])
df_result['token2'] = df_result.token.apply(lambda x: x[1])
```

```python
df_result.head()
```

```python
df_result.rename(columns={'origin':'유저 발화', 'token1':'토큰', 'token2':'품사'}, inplace=True)
```

```python
df_result = df_result[['유저 발화', '토큰', '품사']]
```
