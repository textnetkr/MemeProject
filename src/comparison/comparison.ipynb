# okt

```python
from konlpy.tag import Okt
```

```python
okt = Okt()
```

```python
"""
근데 조금 궁금한 게 있는데 연봉 3천과 6천을 월 실수량으로 바꾸면 어느 정도인 줄
알 수 있을까요?
그 정도 되나요? 60
어느 정도
6천 만원이요 얼마죠?
450에 아홉시 퇴근..
"""

okt.phrases('근데 조금 궁금한 게 있는데 연봉 3천과 6천을 월 실수량으로 바꾸면 어느 정도인 줄')
```

```python
df['pos'] = df.content.swifter.apply(lambda x: okt.pos(str(x)))
```

```python
df.head()
```

##### change shape

```python
result = []
for row in df.iterrows():
  for i in row[1]['pos']:
    temp = []
    temp.append(row[1]['content'])
    temp.append(i)
    result.append(temp)
```

```python
pprint(result[:10])
```

```python
df_okt = pd.DataFrame(result, columns=['origin', 'token'])
```

```python
df_okt['token1'] = df_okt.token.apply(lambda x: x[0])
df_okt['token2'] = df_okt.token.apply(lambda x: x[1])
```

```python
df_okt = df_okt[['origin', 'token1', 'token2']]
df_okt.rename(columns={'origin':'유저 발화', 'token1':'토큰', 'token2':'품사'}, inplace=True)
```

```python
df_okt.head()
```

```python
# import config
from konlpy.tag import Mecab
```

```python
m = Mecab()
```

```python
help(m)
```

```python
df['mecab'] = df.content.swifter.apply(lambda x: m.pos(str(x)))
```

```python
df.head()
```

##### change shape

```python
result = []
for row in df.iterrows():
  for i in row[1]['mecab']:
    temp = []
    temp.append(row[1]['content'])
    temp.append(i)
    result.append(temp)
```

```python
pprint(result[:10])
```

```python
df_mecab = pd.DataFrame(result, columns=['origin', 'token'])
```

```python
df_mecab['token1'] = df_mecab.token.apply(lambda x: x[0])
df_mecab['token2'] = df_mecab.token.apply(lambda x: x[1])
```

```python
df_mecab = df_mecab[['origin', 'token1', 'token2']]
df_mecab.rename(columns={'origin':'유저 발화', 'token1':'토큰', 'token2':'품사'}, inplace=True)
```
