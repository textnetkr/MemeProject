```python
import urllib3
import json
import pandas as pd
```

```python
# openApiURL = "http://aiopen.etri.re.kr/demo/WiseNLU" # 일반 문어 처리할 때
openApiURL = "https://aiopen.etri.re.kr/demo/WiseNLU_spoken"#  구어 처리할 때

# openApiURL = "http://aiopen.etri.re.kr:8000/WiseNLU" # 일반 문어 처리할 때
# openApiURL = "http://aiopen.etri.re.kr:8000/WiseNLU_spoken" #  구어 처리할 때
```

```python
# 분석기 종류 #
parsing = "dparse" # 구문 분석기, 문어만 가능
pos = "morp" # 형태소 분석기, 문어/구어 다 가능 
ner = "ner" # 개체명 인식기, 문어/구어 다 가능 
wsd = "wsd" # 동음이의어 분석기, 문어만 가능 

accessKey = "e5cb4039-cf69-4d76-9a67-6698eff0021f" # ETRI 홈페이지에서 api 키를 미리 신청받아야 합니다.
analysisCode = parsing # 원하는 분석기 종류를 골라서 쓰면 됩니다.

text = "윤동주(尹東柱, 1917년 12월 30일 ~ 1945년 2월 16일)는 한국의 독립운동가, 시인, 작가이다."
        

requestJson = {
    "access_key": accessKey,
    "argument": {
        "text": text,
        "analysis_code": analysisCode
    }
}

http = urllib3.PoolManager()
response = http.request( # 결과는 json 형식으로 들어옴
    "POST",
    openApiURL,
    headers={"Content-Type": "application/json; charset=UTF-8"},
    body=json.dumps(requestJson)
)

print("[responseCode] " + str(response.status))
print("[responBody]")
print(str(response.data,"utf-8"))
```

##### WordExtractor

```python
# WordExtractor
from soynlp.word import WordExtractor

word_extractor = WordExtractor(
    min_frequency=10,
    min_cohesion_forward=0.05, 
    min_right_branching_entropy=0.0
)

word_extractor.train(u)  # list of str or like
# word_extractor.save(cfg.soynlp)    # 모델 저장
```

```python
words = word_extractor.extract()
print(len(words))
print(words['중간고사'])
```

```python
from soynlp.tokenizer import LTokenizer

cohesion_score = {word:score.cohesion_forward for word, score in words.items()}
tokenizer = LTokenizer(scores=cohesion_score)

sent = "홍대 치킨집이 선행했대. 레게노다. 돈쭐내주자!"
print(tokenizer.tokenize(sent))
```

##### add dict

```python
cohesion_score["선행"] = 1.0
cohesion_score["레게노"] = 1.0
cohesion_score["돈쭐"] = 1.0
```
