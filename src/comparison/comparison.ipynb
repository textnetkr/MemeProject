# Etri

```python
import urllib3
import json
import pandas as pd
```

```python
# openApiURL = "http://aiopen.etri.re.kr/demo/WiseNLU" # 일반 문어 처리할 때
openApiURL = "https://aiopen.etri.re.kr/demo/WiseNLU_spoken"#  구어 처리할 때

# openApiURL = "http://aiopen.etri.re.kr:8000/WiseNLU" # 일반 문어 처리할 때
# openApiURL = "http://aiopen.etri.re.kr:8000/WiseNLU_spoken" #  구어 처리할 때
```

```python
# 분석기 종류 #
parsing = "dparse" # 구문 분석기, 문어만 가능
pos = "morp" # 형태소 분석기, 문어/구어 다 가능 
ner = "ner" # 개체명 인식기, 문어/구어 다 가능 
wsd = "wsd" # 동음이의어 분석기, 문어만 가능 

accessKey = "e5cb4039-cf69-4d76-9a67-6698eff0021f" # ETRI 홈페이지에서 api 키를 미리 신청받아야 합니다.
analysisCode = parsing # 원하는 분석기 종류를 골라서 쓰면 됩니다.

text = "윤동주(尹東柱, 1917년 12월 30일 ~ 1945년 2월 16일)는 한국의 독립운동가, 시인, 작가이다."
        

requestJson = {
    "access_key": accessKey,
    "argument": {
        "text": text,
        "analysis_code": analysisCode
    }
}

http = urllib3.PoolManager()
response = http.request( # 결과는 json 형식으로 들어옴
    "POST",
    openApiURL,
    headers={"Content-Type": "application/json; charset=UTF-8"},
    body=json.dumps(requestJson)
)

print("[responseCode] " + str(response.status))
print("[responBody]")
print(str(response.data,"utf-8"))
```

# Okt

```python
from konlpy.tag import Okt
```

```python
okt = Okt()
```

```python
"""
근데 조금 궁금한 게 있는데 연봉 3천과 6천을 월 실수량으로 바꾸면 어느 정도인 줄
알 수 있을까요?
그 정도 되나요? 60
어느 정도
6천 만원이요 얼마죠?
450에 아홉시 퇴근..
"""

okt.phrases('근데 조금 궁금한 게 있는데 연봉 3천과 6천을 월 실수량으로 바꾸면 어느 정도인 줄')
```

```python
df['pos'] = df.content.swifter.apply(lambda x: okt.pos(str(x)))
```

```python
df.head()
```

##### change shape

```python
result = []
for row in df.iterrows():
  for i in row[1]['pos']:
    temp = []
    temp.append(row[1]['content'])
    temp.append(i)
    result.append(temp)
```

```python
pprint(result[:10])
```

```python
df_okt = pd.DataFrame(result, columns=['origin', 'token'])
```

```python
df_okt['token1'] = df_okt.token.apply(lambda x: x[0])
df_okt['token2'] = df_okt.token.apply(lambda x: x[1])
```

```python
df_okt = df_okt[['origin', 'token1', 'token2']]
df_okt.rename(columns={'origin':'유저 발화', 'token1':'토큰', 'token2':'품사'}, inplace=True)
```
