# Data Load

```python
import sys
sys.path.append('/content/drive/MyDrive/MemeProject/src/comparison')
print(sys.path)


import swifter
import config
from pshmodule.utils import filemanager as fm
```

```python
df = fm.load(config.origin_data)
```

```python
df = df[['content']]
```

```python
df.head()
```

# Tokenizer

# okt

```python
from konlpy.tag import Okt
```

```python
okt = Okt()
```

```python
df['pos'] = df.content.swifter.apply(lambda x: okt.pos(str(x)))
```

# mecab

```python
from konlpy.tag import Mecab
```

```python
m = Mecab()
```

```python
df['mecab'] = df.content.swifter.apply(lambda x: m.pos(str(x)))
```

# pecab

```python
!pip3 install pecab
```

```python
from pecab import PeCab
```

```python
p = PeCab()
```

```python
df['pecab'] = df.content.swifter.apply(lambda x: p.pos(str(x)))
```
