Tag Mapping

Okt

```python
def okt(pos):
  result = ""
  if pos in ['NN', 'NNP', 'NR', 'NP']:
    result = "Noun" # 명사
  elif pos == "VV":
    result = "Verb" # 동사
  elif pos == "VA":
    result = "Adjective" # 형용사
  elif pos == "MM":
    result = "Determiner" # 관형사
  elif pos in ['MAG', 'MAJ']:
    result = "Adverb" # 부사
  elif pos == "IC":
    result = "Exclamation" # 감탄사
  elif pos == "JO":
    result = "Josa" # 조사
  elif pos == "EO":
    result = "Eomi" # 어미
  return result
```

```python
user_dict_3['okt_pos'] = user_dict_3.pos.swifter.apply(okt)
```

Mecab

```python
def mecab(pos):
  result = ""
  if pos == "NN":
    result = "NNG" # 일반 명사
  elif pos == "NNP":
    result = "NNP" # 고유 명사
  elif pos == "NR":
    result = "NR" # 수사
  elif pos == "NP":
    result = "NP" # 대명사
  elif pos == "VV":
    result = "VV" # 동사
  elif pos == "VA":
    result = "VA" # 형용사
  elif pos == "MM":
    result = "MM" # 관형사
  elif pos == "MAG":
    result = "MAG" # 일반부사
  elif pos == "MAJ":
    result = "MAJ" # 접속부사
  elif pos == "IC":
    result = "IC" # 감탄사
  elif pos == "JO":
    result = "JKS" # 조사
  elif pos == "EO":
    result = "EF" # 조사
  return result
```

```python
user_dict_3['mecab_pos'] = user_dict_3.pos.swifter.apply(mecab)
```

```python
# trie 구조로 속도 개선

# 증강 할 유저 발화 86,141건 중 평균 559 중 미만 데이터 29,489건
result = []
for k, v in tqdm(enumerate(df_under_average.iterrows())):
  # 원본
  result.append([v[1]['g_num'], v[1]['u_num'], v[1]['material'], v[1]['speech'], v[1]['u'], v[1]['arg1'], v[1]['arg_op1'], v[1]['arg_op2'], '', v[1]['label']])

  input = [f"{i[0]}, {i[1]}" for i in kiwi.tokenize(v[1]['u'].replace(' ', '')) if i[1] in morpheme]
  check = sum([i in input for i in sum(udict, [])])
  # input 토큰 iterable
  for i in input:
    # trie 검색
    if trie.search(i):
      replace_set = list(set(df_userdict[df_userdict.formal == i.split(',')[0]].lemma.tolist()))
      if len(replace_set) > 1:
        # 교체될 단어가 문장의 첫 어절이면 스페이스를 뒤만.
        # 교체될 단어가 문장의 첫 어절이 아니면 스페이스를 앞뒤로.
        for kk, vv in enumerate(replace_set):
          replace_set_copy = replace_set.copy()
          replace_set_copy.pop(kk)

          for r in replace_set_copy:
            # 어절이 맨 앞인 경우
            if str(v[1]['u']).find(str(vv)) == 0 and str(v[1]['u']) != str(v[1]['u']).replace(str(vv)+" ", str(r)+" "):              
              result.append([v[1]['g_num'], v[1]['u_num'], v[1]['material'], v[1]['speech'], str(v[1]['u']).replace(str(vv)+" ", str(r)+" "), v[1]['arg1'], v[1]['arg_op1'], v[1]['arg_op2'], str(r), v[1]['label']])
            # 어절이 중간에 있는 경우
            elif str(v[1]['u']).find(str(vv)) > 0 and str(v[1]['u']) != str(v[1]['u']).replace(" "+str(vv)+" ", " "+r+" "):
              result.append([v[1]['g_num'], v[1]['u_num'], v[1]['material'], v[1]['speech'], str(v[1]['u']).replace(" "+str(vv)+" ", " "+str(r)+" "), v[1]['arg1'], v[1]['arg_op1'], v[1]['arg_op2'], str(r), v[1]['label']])
            # 조사는 어절의 뒤에 붙기 때문에 목적어를 찾아 뒤에 말에 조사가 있는지를 판단해서 ex) 올림픽 은메달도 -> 올림픽 은메달 x
            for s in stop_list:
              if str(v[1]['u']).find(str(vv)+s+" ") >= 0:
                # 교체할 말이 있는 경우 말이 중복이 생겨 매끄럽지 않기 때문에 제외
                if str(v[1]['u']).find(str(r)+s+" ") < 0:
                  temp = str(v[1]['u']).replace(str(vv)+s+" ", p.jongsung_rep(str(r), s)+" ").split() # 조사 결합 시 종성 체크 후 맞는 조사를 붙여 자연스러운 문장 추가
                  temp_result = " ".join(list(dict.fromkeys(temp)))
                  result.append([v[1]['g_num'], v[1]['u_num'], v[1]['material'], v[1]['speech'], temp_result, v[1]['arg1'], v[1]['arg_op1'], v[1]['arg_op2'], str(r), v[1]['label']])
```

```python
from jamo import h2j, j2hcj

def get_jongsung_TF(sample_text):
    sample_text_list = list(sample_text)
    last_word = sample_text_list[-1]
    last_word_jamo_list = list(j2hcj(h2j(last_word)))
    last_jamo = last_word_jamo_list[-1]

    jongsung_TF = "T"

    if last_jamo in ['ㅏ', 'ㅑ', 'ㅓ', 'ㅕ', 'ㅗ', 'ㅛ', 'ㅜ', 'ㅠ', 'ㅡ', 'ㅣ', 'ㅘ', 'ㅚ', 'ㅙ', 'ㅝ', 'ㅞ', 'ㅢ', 'ㅐ,ㅔ', 'ㅟ', 'ㅖ', 'ㅒ']:
        jongsung_TF = "F"

    return jongsung_TF
```

```python
paths = {
    'NNG.csv':'NNG',
    'NNP.csv':'NNP',
    'NR.csv':'NR',
    'NP.csv':'NP',
    'VV.csv':'VV',
    'VA.csv':'VA',
    'MM.csv':'MM',
    'MAG.csv':'MAG',
    'MAJ.csv':'MAJ',
    'IC.csv':'IC',
    'J.csv':'JO',
}
```

```python
for p in paths:
  with open(f'/tmp/mecab-ko-dic-2.1.1-20180720/{p}', "r", encoding='utf-8') as f:
    user_dict = f.readlines()

  for i in user_dict_3[user_dict_3.mecab_pos == paths[p]].iterrows():
    jongsung_TF = get_jongsung_TF(str(i[1]['lemma']))
    line = '{},*,*,*,{},*,{},{},*,*,*,*,*\n'.format(i[1]['lemma'], paths[p], jongsung_TF, i[1]['lemma'])
    user_dict.append(line)
  
  with open(f'/tmp/mecab-ko-dic-2.1.1-20180720/{p}', 'w', encoding='utf-8') as f:
    for line in user_dict:
      f.write(line)

  user_dict = []
```
