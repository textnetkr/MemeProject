```python
df = fm.load(cfg.origin_ref)
```

```python
df.head()
```

```python
df_check = df[['g_num', 'u']]
```

```python
df_check.drop_duplicates(inplace=True)
```

```python
df_check.reset_index(drop=True, inplace=True)
```

```python
df_check.head(13)
```

# Labeling

```python
df_check['label'] = df_check.g_num.swifter.apply(lambda x: x-1)
```

```python
df_check.rename(columns={'u':'content'}, inplace=True)
```

```python
df_check = df_check[['content', 'label']]
```
