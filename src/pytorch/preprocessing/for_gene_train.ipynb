# Save

```python
import json
```

```python
temp_dict = [{"Text": row['content'].strip(), "Completion": row['label']} for _, row in df.iterrows()]
with open('/content/drive/MyDrive/MemeProject/data/train_for_clova.jsonl', 'w', encoding='utf-8') as f:
    for line in temp_dict:
        json_record = json.dumps(line, ensure_ascii=False)
        f.write(json_record + '\n')
```
